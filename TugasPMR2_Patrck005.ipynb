```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = '/content/onlinefoods.csv'
df = pd.read_csv(dataset)
df
```

```python
# @title Gender

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('Gender').size().plot(kind='barh', color=sns.palettes.mpl_palette('Set2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

Jadi setelah melihat hasil bar plot horizontal diatas , dapat diketahui bahwa jumlah pemesanan makanan online lebih banyak dilakukan oleh laki" daripada perempuan


```python
from matplotlib import pyplot as plt

# Plot histogram dengan warna yang diinginkan (misalnya, hijau)
df['Age'].plot(kind='hist', bins=20, title='Age', color='pink')

# Menghilangkan garis tepi atas dan kanan dari plot
plt.gca().spines[['top', 'right']].set_visible(False)

# Menampilkan plot
plt.show()
```


Jadi setelah melihat hasil histogram diatas , dapat diketahui bahwa jumlah pemesanan makanan online lebih banyak dilakukan oleh orang berumur 22-24 tahun
